Using the latitude and longitude from the DataFrame above and put it through Foursquare API query to obtain cafes (chosen point of interest) around all the bike stations. 

```python

foursquare_cafe = df
foursquare_cafe['F_listOfCafes'] = ''
foursquare_cafe['F_listOfRatings'] = ''

#for n in range(3):
for n in range(len(df)):
    try:
        latitude = df.at[n, 'latitude']
        longitude = df.at[n, 'longitude']

        foursquare_url = f"https://api.foursquare.com/v3/places/search?query=cafe&ll={latitude}%2C{longitude}&radius=1000&fields=name%2Crating"
        headers = {"accept": "application/json",
                   "Authorization": "fsq3NScpiz3FYehTY/n+gI+1wwaBdmMpmaMbtX4vAOjcAi4="}

        foursquare_response = requests.get(foursquare_url, headers=headers) 
        foursquare_response_json = json.loads(foursquare_response.text)
        foursquare = foursquare_response_json['results']
        pd_foursquare = pd.DataFrame(foursquare)
        cafe_list = pd_foursquare['name'].tolist()
        rating_list = pd_foursquare['rating'].tolist()
        
        foursquare_cafe.at[n, 'F_listOfCafes'] = cafe_list
        foursquare_cafe.at[n, 'F_numberOfCafes'] = len(cafe_list)
        foursquare_cafe.at[n, 'F_listOfRatings'] = rating_list
        foursquare_cafe.at[n, 'F_averageRatings'] = sum(rating_list)/len(rating_list)
        
    except:
        pass
    
foursquare_cafe
```

Rename the columns for readability. 

```python
foursquare_cafe_final = foursquare_cafe.rename(columns={
    'name': 'bikeStationName',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'total_bike_slots': 'totalBikeSlots'
})
foursquare_cafe_final
```

Data Cleaning: Handling null values

Notice that there are some empty lists in listOfCafes and the numberOfCafes are null values. Assume that if listOfCafes is empty which means there are no cafes within 1km. Replace null values with 0.

Looking at F_listOfRatings, there are a lot of cafes hat don't actually have a rating, which are the null values. Since there are not enough information, it is best to keep it as is instead of replacing the null values. Same goes to the F_averageRatings column due to lack of information. 

```python
foursquare_cafe_final['F_numberOfCafes'] = foursquare_cafe_final['F_numberOfCafes'].fillna(0)
foursquare_cafe_final
```
