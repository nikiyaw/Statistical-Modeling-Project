```python
import requests
import pandas as pd
from pprint import pprint
import json
```

Using the latitude and longitude from the DataFrame above and put it through YELP API query to obtain cafes (chosen point of interest) around all the bike stations.

```python
yelp_cafe = df
yelp_cafe['Y_listOfCafes'] = ''
yelp_cafe['Y_listOfRatings'] = ''

#for n in range(3):
for n in range(len(df)):
    try:
        latitude = df.at[n, 'latitude']
        longitude = df.at[n, 'longitude']
        
        yelp_url = f"https://api.yelp.com/v3/businesses/search?latitude={latitude}&longitude={longitude}&term=cafe&radius=1000&sort_by=best_match&limit=20"
        headers = {"accept": "application/json",
                   "Authorization": "Bearer kRiwQHGufRBXSekWwL7gjfg6qLctefkzSvAmrcDpj_runTiC6sHj_B6OjvIcpJovymJXEeUXhwxfewAyKsZnsdpcVsAxbDAzNeWrBtepp7TwpOf7CakwVhOXmlJ6ZHYx"}
 
        yelp_response = requests.get(yelp_url, headers=headers)
        yelp_response_json = json.loads(yelp_response.text)
        yelp = yelp_response_json['businesses']
        pd_yelp = pd.DataFrame(yelp)
        cafe_list = pd_yelp['name'].tolist()
        rating_list = pd_yelp['rating'].tolist()

        yelp_cafe.at[n, 'Y_listOfCafes'] = cafe_list
        yelp_cafe.at[n, 'Y_numberOfCafes'] = len(cafe_list) 
        yelp_cafe.at[n, 'Y_listOfRatings'] = rating_list
        yelp_cafe.at[n, 'Y_averageRatings'] = sum(rating_list)/len(rating_list)
              
    except:
        pass
    
yelp_cafe
```

Rename the columns for readability.

```python
yelp_cafe_final = yelp_cafe.rename(columns={
    'name': 'bikeStationName',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'total_bike_slots': 'totalBikeSlots',
})
yelp_cafe_final
```

Data Cleaning: Handling null values

Notice that there are some empty lists in Y_listOfCafes and the Y_numberOfCafes that are null values. 
Assume that if Y_listOfCafes is empty which means there are no cafes within 1km. 
Replace null values with 0. 

Y_averageRatings only produces null values when Y_listofRatings is empty. 
Keeping it as null would be more appropriate because if we replace it by 0, it means that average ratings of cafes around that area is actually 0 (worst service). 

```python
yelp_cafe_final['Y_numberOfCafes'] = yelp_cafe_final['Y_numberOfCafes'].fillna(0)
yelp_cafe_final
```
